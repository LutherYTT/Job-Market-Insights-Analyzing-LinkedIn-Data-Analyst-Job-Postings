```python
import pandas as pd

def preprocess_csv(file_path):
    # Load the CSV file
    df = pd.read_csv(file_path)

    # Remove rows where 'skill(1)' is empty or NaN
    df_filtered = df[df['skill(1)'].notna() & (df['skill(1)'] != '')]

    return df_filtered

def combine_skills(df):
    # Replace "和" with "、" in skill(1)
    df['skill(1)_clean'] = df['skill(1)'].str.replace('和', '、', regex=False)

    # Fill NaN values in skill(2) with an empty string
    df['skill(2)'] = df['skill(2)'].fillna('')

    # Combine skill(1)_clean and skill(2) with "、" as separator
    df['skill'] = df['skill(1)_clean'] + '、' + df['skill(2)']

    # Optional: If you want to remove duplicate separators like "、、"
    df['skill'] = df['skill'].str.replace('、+', '、', regex=True)

    # Drop the temporary column if you want
    df.drop(columns=['skill(1)_clean'], inplace=True, errors='ignore')

    return df
```

```python
if __name__ == "__main__":
    preprocess_df = preprocess_csv('linkedin_raw_data.csv')
    df = combine_skills(preprocess_df)

    df.to_csv("preprocessed_linkedin_data.csv", index=False)
```
